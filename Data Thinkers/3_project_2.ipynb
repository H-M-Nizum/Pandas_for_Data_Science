## 2. Perform Data Cleaning [Replace '?' with NaN]

```python
# Check number of '?'
df.isin(['?']).sum()
```

```python
import numpy as np
df['workclass'] = df['workclass'].replace('?', np.nan)
df['occupation'] = df['occupation'].replace('?', np.nan)
df['native-country'] = df['native-country'].replace('?', np.nan)
```

```python
df.head()
```

```python
df.isin(['?']).sum()
```

```python
df.isnull().sum()
```

```python
# Data Visualization using heatmap
sns.heatmap(df.isnull())
```

## 3. Drop all the missing values

```python
df.shape
```

```python
# Pandas dropna() method allows the user to analyze and drop Rows/Columns with Null values in different ways. 
# Syntax: DataFrameName.dropna(axis=0, how=’any’, thresh=None, subset=None, inplace=False)

df.dropna(how='any', inplace=True)
df.shape
```

## 4. Check for Duplicate data and drop them

```python
# Syntax: DataFrame.drop_duplicates(subset=None, keep=’first’, inplace=False)
data = df.drop_duplicates()
```
