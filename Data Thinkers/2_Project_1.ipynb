# New Salary DataSet : https://www.kaggle.com/datasets/kaggle/sf-salaries

```python
new_df = pd.read_csv("E:\Github Advance Python\Pandas_for_Data_Science\DS\Salaries.csv")
new_df
```

## 2. Replace 'Not Provided' in EmployeeName column to NaN

```python
import numpy as np
new_df['EmployeeName'] = new_df['EmployeeName'].replace('Not provided', np.nan)
```

```python
new_df['EmployeeName']
```

## 3. Drop the Rows having 3 missing values

```python
new_df.isnull().sum()
```

```python
new_df.drop(new_df[new_df.isnull().sum(axis=1) == 3].index, axis=0)
```

## 5. Find Average TotalPay of All Employee per year

```python
# Group by 'Year' and calculate the mean of 'TotalPay'
average_total_pay_per_year = new_df.groupby('Year')['TotalPay'].mean()

# Display the result
print(average_total_pay_per_year)
```
